```python
def perceptron(x,y):
  x=x.reshape(-1,2)
  x=np.insert(x,0,1,axis=1)
  w=np.ones(x.shape[1])
  lr=0.1
  for i in range(1000):
    j=np.random.randint(0,100)
    y_hat= step(np.dot(x[j],w))  #dot product
    w=w + lr*(y[j]-y_hat)*x[j]
  return w[0],w[1:]
```
